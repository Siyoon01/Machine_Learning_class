```python
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/ML/weather.csv", index_col=0)
df  # print(df) 와 동일
```

```python
# RainTomorrow 컬럼 삭제

df = df.drop('RainTomorrow', axis=1)
df
```

```python
# 결측치 확인

df.isnull().sum()
```

```python
# 결측치 제거

df = df.dropna()
df.isnull().sum()
```

```python
# 컬럼명 확인

df.columns
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoders = {}
for column in df.columns:
    label_encoders[column] = LabelEncoder()
    df[column] = label_encoders[column].fit_transform(df[column])
```

```python
# 입력과 출력 분리

y = df['Rainfall']
y.head()
```

```python
# 입력과 출력 분리

X = df.drop('Rainfall',axis=1)
X.head()
```

```python
# 훈련과 테스트 데이터 분리

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test=train_test_split(
    X, y, test_size=0.2, shuffle=True, random_state=12
)

print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
# 모델 구성, 훈련, 예측 데이터 생성, 평가

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

model = LinearRegression()         # 모델 구성
model.fit(X_train, y_train)        # 모델 훈련
ly_preds = model.predict(X_test)   # 예측데이터 생성
print(mean_squared_error(y_test, ly_preds))  #평가
```

```python
# 예측값과 실제 값 차이를 산접도로 표시, 노란색 예측

import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.scatter(X_test['Evaporation'], y_test, label='y_test')
plt.scatter(X_test['Evaporation'], ly_preds, c='y', label='ly_preds')
#plt.plot(X_train, lr.predict(X_train), color='blue')
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()
model.fit(X_train, y_train)

from sklearn.metrics import mean_squared_error

ry_preds = model.predict(X_test)
print('평균제곱근오차', mean_squared_error(ry_preds, y_test))
```

```python
plt.figure(figsize=(10,5))
plt.scatter(X_test['Evaporation'], y_test, label='y_test')
plt.scatter(X_test['Evaporation'], ry_preds, c='orange', label='ry_preds')

plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor

model = DecisionTreeRegressor()
model.fit(X_train, y_train)

from sklearn.metrics import mean_squared_error

dy_preds = model.predict(X_test)
print('평균제곱근오차', mean_squared_error(dy_preds, y_test))
```

```python
plt.figure(figsize=(10,5))
plt.scatter(X_test['Evaporation'], y_test, label='y_test')
plt.scatter(X_test['Evaporation'], dy_preds, c='g', label='dy_preds')
plt.show()
```
